```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Dell\Documents\AI_EDA\AI_Impact_Student_Life_2026.csv")
```

```python
fig,ax = plt.subplots(1,2,figsize=(6,3))
df_before =df['GPA_Baseline'].plot(kind='hist',bins=20,edgecolor='black',ax=ax[0])
df_after=df['GPA_Post_AI'].plot(kind='hist',bins=20,edgecolor='black',ax=ax[1])
ax[0].axvline(df['GPA_Baseline'].median(),color='y',linestyle='--',label='Median')
ax[1].axvline(df['GPA_Post_AI'].median(),color='y',linestyle='--',label='Median')
ax[1].legend()
ax[1].set_ylabel('')
rt=ax[0].set_ylim(0,190)
ax[1].set_ylim(auto=True)
ax[0].set_title('GPA Before AI')
ax[1].set_title('GPA After AI')
ax[0].set_ylabel('Number of Students')
ax[1].legend(bbox_to_anchor=(1,1))
```

```python
df_Corr_Analysis = df[["Task_Frequency_Daily","GPA_Post_AI"]]
```

```python
sns.regplot(df_Corr_Analysis,x='Task_Frequency_Daily',y='GPA_Post_AI',scatter_kws={'alpha':0.5})
sns.despine()
plt.ylabel('GPA Post AI')
plt.xlabel('Task Frequency Daily')
plt.title('Scores post AI v/s the Times AI used Daily')
```

```python
plt.figure(figsize=(5,3))
sns.boxplot(data=df,x='GPA_Post_AI',y='Primary_AI_Tool',color='grey')
sns.despine()

plt.title('Scores after using Different AIs')
plt.ylabel('')
plt.xlabel('Scores')
plt.axvline(df['GPA_Post_AI'].median(),color='y',linestyle='--',label='Median')
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

```python
sns.scatterplot(data=df,x='Age',y='Task_Frequency_Daily')
sns.despine()
plt.title('Ages associated with daily AI Activities')
plt.ylabel('Task Daily Frequency')
plt.xlabel('Age')
```
