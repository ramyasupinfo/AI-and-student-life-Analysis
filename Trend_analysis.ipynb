```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Dell\Documents\AI_EDA\AI_Impact_Student_Life_2026.csv")
```

```python
df
```

```python
df_pivot = df.pivot_table(index='Age',aggfunc='size',columns='Primary_AI_Tool')
df_pivot.loc['Total'] = df_pivot.sum().sort_values(ascending=False)
df_pivot = df_pivot[df_pivot.loc['Total'].sort_values(ascending=False).index]
df_pivot = df_pivot.drop('Total')
```

```python
df_pivot
```

```python
df_pivot =df_pivot.reset_index()
```

```python
df_pivot = df_pivot.set_index('Age')
```

```python
df_pivot=df_pivot.iloc[:,:2]
```

```python
df_pivot.plot(kind='line',marker='o')
plt.title('Top 2 AI Tools used across various age groups')
plt.xlabel('Age')
plt.ylabel('Count of users')
```
